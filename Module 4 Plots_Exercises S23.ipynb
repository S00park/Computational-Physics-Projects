### Exercise 1

$$x(t) = a \cos (nt) \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \  (1)$$
$$y(t) = b \cos (mt + \phi) \ \ \ \ \ \ \ \ \ \ (2)$$

(e) Experiment with different values of the constants `a`, `b`, `n`, `m`, and `phi`. Re-run your cell and see how the plot changes with these different constants.

```python
# Exercise 1

a=1
b=1
n=1
m=1

phi=np.pi/2

t = np.linspace(0, 2*(np.pi),101)


x = (np.cos(n*t))
y = (np.cos(m*t+phi))

plt.plot(x,y) 
plt.axis('equal')
plt.xlabel('x',size=15) 
plt.ylabel('y',size=15)


plt.show()
```

## Exercise 5, A Valentine's Day Plot

In the equations below, $x$ is a function of $z$, and $y$ is also a function of $z$. In other words, for any given value of $z$, there is a corresponding pair of numbers $(x(z), y(z))$. These equations are called **parametric equations**, with $z$ being the parameter. (Exercise 1 is another example of parametric equations, where the time $t$ is the parameter.)

$$x(z) = 16 sin^3(z)$$
$$y(z) = 13 cos(z)-5 cos(2z)-2 cos(3z)- cos(4z)$$


plot $y$ vs $x$ for $z$ ranging from 0 to 2$\pi$ with 100 points and plot it in red, of course. Use a line width of 10 (lw = 10). Make the figure a square using the command `plt.figure(figsize=(8,8))`. 

Add a title to your plot using this command: `plt.title('Happy Valentine\'s Day!')` (Please note that the backslash here is used to indicate that the apostrophe following it is truly an apostrophe, and not the end of the string variable before it.)

```python
# Exercise 5
z=np.linspace(0,2*np.pi,100)


x=16*np.sin(z)**3
y=13*np.cos(z)-5*np.cos(2*z)-2*np.cos(3*z)-np.cos(4*z)

plt.plot(x, y,markersize=2, linewidth=10, color="red") 
plt.title('Happy Valentine\'s Day!')
plt.axis('equal')
plt.xlabel('x', fontsize=12)   
plt.ylabel('y', fontsize=12)
plt.figure(figsize=(8,8))

plt.show()
```
